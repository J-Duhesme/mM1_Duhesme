```python
#trouver token qui dépendent de pain

def extract_dependents(file_path, output_path, target_lemmas):
    with open(file_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    sentences = []
    sentence = []
    for line in lines:
        line = line.strip()
        if line == "":
            if sentence:
                sentences.append(sentence)
                sentence = []
        elif not line.startswith("#"):
            sentence.append(line)
    if sentence:
        sentences.append(sentence)

    result = []

    for sent in sentences:
        token_list = []
        for line in sent:
            parts = line.split('\t')
            if len(parts) < 8 or '-' in parts[0] or '.' in parts[0]:
                continue
            token_list.append({
                "index": int(parts[0]),
                "word": parts[1],
                "lemma": parts[2].lower(),
                "head": int(parts[6])
            })

        target_tokens = [t for t in token_list if t["lemma"] in target_lemmas]
        if not target_tokens:
            continue

        target_indices = {t["index"] for t in target_tokens}

        dependents = [
            t for t in token_list
            if t["head"] in target_indices and t["lemma"] not in target_lemmas
        ]

        if dependents:
            sorted_deps = sorted(dependents, key=lambda x: x["index"])
            result.append(" ".join(t["word"] for t in sorted_deps))
        else:
            result.append("-")

    with open(output_path, 'w', encoding='utf-8') as f_out:
        for line in result:
            f_out.write(line + "\n")

    print(f"Saved: {output_path}")
```

```python
for i in range(3):
    extract_dependents(docs[i], outputs_s[i], {"pain", "pains"})
    extract_dependents(docs[i], outputs_a[i], {"painful"})
    extract_dependents(docs[i], outputs_y[i], {"painfully"})
```

```python
#trouver token dont pain dépend

def extract_heads_of_target_lemmas(file_path, output_txt_path, target_lemmas):
    with open(file_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    sentences = []
    sentence = []
    for line in lines:
        line = line.strip()
        if line == "":
            if sentence:
                sentences.append(sentence)
                sentence = []
        elif not line.startswith("#"):
            sentence.append(line)
    if sentence:
        sentences.append(sentence)

    result = []

    for sent in sentences:
        token_list = []
        index_map = {}

        for line in sent:
            parts = line.split('\t')
            if len(parts) < 8 or '-' in parts[0] or '.' in parts[0]:
                continue
            idx = int(parts[0])
            word = parts[1]
            lemma = parts[2].lower()
            head = int(parts[6])

            token_data = {
                "index": idx,
                "word": word,
                "lemma": lemma,
                "head": head
            }
            token_list.append(token_data)
            index_map[idx] = token_data

        for token in token_list:
            if token["lemma"] in target_lemmas:
                head_idx = token["head"]
                if head_idx in index_map:
                    result.append(index_map[head_idx]["word"])
                else:
                    result.append("-")

    with open(output_txt_path, 'w', encoding='utf-8') as f_out:
        for line in result:
            f_out.write(line + "\n")

    print(f"Saved: {output_txt_path}")

```

```python
for i in range(3):
    extract_heads_of_target_lemmas(docs[i], o_s[i], {"pain", "pains"})
    extract_heads_of_target_lemmas(docs[i], o_a[i], {"painful"})
    extract_heads_of_target_lemmas(docs[i], o_y[i], {"painfully"})
```
